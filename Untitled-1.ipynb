# Document Classification

Le projet consiste à développer un modèle de classification automatique de documents en utilisant des techniques modernes de traitement du langage naturel (NLP). L’objectif est de prédire la catégorie thématique d’un texte en fonction de son contenu.

Pour ce faire, nous utilisons le BBC Full-Text Dataset, un corpus de documents collectés à partir du site de la BBC. Ce dataset est largement utilisé dans la recherche pour tester les performances des algorithmes de classification de texte.

***Librairies***

```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# NLTK
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# SpaCy
import spacy

# Scikit-learn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support

# PyTorch & Transformers
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
```

## Lecture des fichiers et création du DataFrame

```python
DATA_DIR = "./data"

# DATA_DIR = "C:\\Users\\ZAYED\\Desktop\\projet ai\\data"
rows = []

for category in os.listdir(DATA_DIR):
    cat_path = os.path.join(DATA_DIR, category)
    if not os.path.isdir(cat_path):
        continue
    for fname in os.listdir(cat_path):
        if not fname.lower().endswith(".txt"):
            continue
        fpath = os.path.join(cat_path, fname)
        try:
            with open(fpath, "r", encoding="utf-8") as f:
                text = f.read()
        except UnicodeDecodeError:
            with open(fpath, "r", encoding="latin-1", errors="ignore") as f:
                text = f.read()
        rows.append({"text": text, "label": category, "filename": fname})


```

```python
df = pd.DataFrame(rows)
print("Total docs:", len(df))
print(df['label'].value_counts())
df.head()
```

## Nettoyage du dataframe et Preprocessing

```python
# nltk
nltk.download('punkt')

nltk.download('punkt_tab')

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
STOP = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(s):
    # 1. Mise en minuscule
    s = s.lower()
    # 2. Retirer urls, caractères spéciaux, nombres
    s = re.sub(r'http\S+', '', s)
    s = re.sub(r'[^a-z\s]', ' ', s)         # garde uniquement les lettres
    s = re.sub(r'\s+', ' ', s).strip()
    
    # 3. Tokenisation
    tokens = nltk.word_tokenize(s)
    
    # 4. Suppression des stopwords + lemmatisation
    tokens = [lemmatizer.lemmatize(tok) for tok in tokens if tok not in STOP]
    
    # 5. Rejoindre en texte nettoyé
    return " ".join(tokens)

# Application
df['text_clean'] = df['text'].astype(str).apply(clean_text)
df['char_len'] = df['text'].str.len()
df['word_count'] = df['text_clean'].str.split().apply(len)

print(df[['text','text_clean']].head())
```

# SpaCy

```python
nlp = spacy.load("en_core_web_sm")

def spacy_preprocess(text):
    doc = nlp(text.lower())  # passe en minuscule
    tokens = []
    for token in doc:
        # Garder seulement les mots utiles
        if not token.is_stop and not token.is_punct and not token.like_num:
            lemma = token.lemma_.strip()   # lemmatisation
            if lemma != "":
                tokens.append(lemma)
    return " ".join(tokens)


df["text_clean"] = df["text"].astype(str).apply(spacy_preprocess)


df["char_len"] = df["text"].str.len()
df["word_count"] = df["text_clean"].str.split().apply(len)

print(df[["text","text_clean"]].head())
```

## Visualisation du dataframe 


```python
# distribution des classes
counts = df['label'].value_counts()
counts.plot(kind='bar', title='Documents par catégorie', figsize=(6,4))
plt.show()

# afficher les mots les plus fréquents par classe (CountVectorizer).
cv = CountVectorizer(stop_words='english', max_features=10000, ngram_range=(1,1))
Xcv = cv.fit_transform(df['text_clean'])
vocab = np.array(cv.get_feature_names_out())

for cat in df['label'].unique():
    idx = df['label']==cat
    s = Xcv[idx].sum(axis=0)  # somme des comptes
    top_n = np.array(s).ravel().argsort()[-15:][::-1]
    print(f"\nTop words for {cat}:")
    print(", ".join(vocab[top_n]))
```

## Train/Test 

```python
RND = 42
X = df['text_clean'].values
y = df['label'].values

X_train, X_tmp, y_train, y_tmp = train_test_split(X, y, test_size=0.30, stratify=y, random_state=RND)
X_val, X_test, y_val, y_test = train_test_split(X_tmp, y_tmp, test_size=0.5, stratify=y_tmp, random_state=RND)

print("Train:", len(X_train), "Val:", len(X_val), "Test:", len(X_test))
```

## Baseline et SVM

```python
tfidf = TfidfVectorizer(ngram_range=(1,2), max_features=10000)  # unigr + bigram
Xtr_tfidf = tfidf.fit_transform(X_train)
Xv_tfidf = tfidf.transform(X_val)
Xt_tfidf = tfidf.transform(X_test)

# Naive Bayes
nb = MultinomialNB()
nb.fit(Xtr_tfidf, y_train)
pred_val_nb = nb.predict(Xv_tfidf)
print("NB val acc:", accuracy_score(y_val, pred_val_nb))
print(classification_report(y_val, pred_val_nb))

# SVM
svm = LinearSVC()
svm.fit(Xtr_tfidf, y_train)
pred_val_svm = svm.predict(Xv_tfidf)
print("SVM val acc:", accuracy_score(y_val, pred_val_svm))
print(classification_report(y_val, pred_val_svm))
```

Les deux modèles sont performants pour ce projet de classification de documents.

NB est un bon choix si tu veux rapidité et simplicité.

SVM reste une alternative robuste et peut être combiné avec GridSearch pour optimiser les hyperparamètres.

Pour améliorer encore la classification :

enlever les mots trop fréquents comme said, year (stopwords spécifiques).

ajouter des n-grams (2-grams) pour capter des expressions caractéristiques (mobile phone, central bank).

éventuellement tester un modèle Transformer si tu veux une performance encore plus fine sur des textes plus longs ou complexes.

```python
y_pred_test = svm.predict(Xt_tfidf)
print("Test acc:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))
```

## GridSearch 

```python
# Pipeline TF-IDF 
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=10000, stop_words='english')),
    ('svm', LinearSVC())
])

# Grille de paramètres à tester
param_grid = {
    'tfidf__max_features': [5000, 10000, 20000],
    'svm__C': [0.1, 1, 10]
}

# GridSearchCV
grid = GridSearchCV(pipeline, param_grid, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=2)
grid.fit(X_train, y_train)

print("Best parameters:", grid.best_params_)
print("Best CV score:", grid.best_score_)

# Évaluer sur test
y_pred_test = grid.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```

## Transformers (BERT)

### Tokeniser

```python
#Encoder les labels (texte -> entier)
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_val_enc   = le.transform(y_val)
y_test_enc  = le.transform(y_test)

label2id = {lab: i for i, lab in enumerate(le.classes_)}
id2label = {i: lab for lab, i in label2id.items()}
print("Label mapping:", label2id)
```

```python
# Tokenizer
MODEL_NAME = "distilbert-base-uncased"
MAX_LENGTH = 256
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
# Dataset torch personnalisé
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=256):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])
        enc = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )
        item = {k: v.squeeze(0) for k,v in enc.items()}
        item["labels"] = torch.tensor(label, dtype=torch.long)
        return item

train_dataset = TextDataset(X_train, y_train_enc, tokenizer, MAX_LENGTH)
val_dataset   = TextDataset(X_val,   y_val_enc,   tokenizer, MAX_LENGTH)
test_dataset  = TextDataset(X_test,  y_test_enc,  tokenizer, MAX_LENGTH)
```

### Modèle 

```python
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=len(le.classes_),
    id2label=id2label,
    label2id=label2id )
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    p, r, f1, _ = precision_recall_fscore_support(labels, preds, average="weighted")
    return {"accuracy": acc, "precision": p, "recall": r, "f1": f1}
```

```python
training_args = TrainingArguments(
    output_dir="bert_bbc_results",
    do_train=True,
    do_eval=True,
    num_train_epochs=2,  
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_steps=50,
    save_steps=100,
    seed=42
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
results = trainer.evaluate()
print(results)
```
